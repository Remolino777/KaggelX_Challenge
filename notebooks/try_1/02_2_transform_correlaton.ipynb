```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import joblib as jb
```

```python
path = r'D:\0_Respaldo\0_Proyectos_2024\ML_proyects\KagelX\KaggelX_Challenge\data\transform_data\clean_data.parquet'
```

```python
data = pd.read_parquet(path)
```

```python
data.head(3)
```

```python
#Result of the correlation analysis the columns will be modelyear , milage, price, engine,accidents.

df = data[['model_year','milage','engine','accident','price']].copy()
```

```python
df.head(3)
```

### Transform categorical

```python
label_encoder = LabelEncoder()
```

```python
#Categorical transform
df['engine_t'] = label_encoder.fit_transform(df['engine'])
df['accident_t'] = label_encoder.fit_transform(df['accident'])
```

```python
# delete columns

del df['engine']

del df['accident']
```

```python
df.head(5
        )
```

### Engeniering

#### Miles by year

```python
df_price = df.pop('price')
```

```python
df1 = df.copy()
```

```python
# Milage by year col
df1['milage_by_year'] = round((df['milage']/(2025 - df['model_year'])),2)
```

```python
df1.head(5)
```

#### -5000 miles and -35miles coulumns

```python
df1['less_5k_millas'] = df['milage'].apply(lambda x: 1 if x < 5000 else 0)
```

```python
df1['less_35k_millas'] = df['milage'].apply(lambda x: 1 if x < 35000 else 0)
```

```python
df1.head(5)
```

```python
df1.describe()
```

#### Concatenar files

```python
df_f = pd.concat([df1,df_price], axis=1)
```
